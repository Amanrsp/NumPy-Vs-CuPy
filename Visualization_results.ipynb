```python
import json
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
with open("drive/My Drive/Colab Notebooks/numpy_results.json", "r") as f:
    numpy_results = json.load(f)

with open("drive/My Drive/Colab Notebooks/cupy_results.json", "r") as f:
    cupy_results = json.load(f)
```

```python
sizes = list(numpy_results.keys())
```

```python
numpy_matrix_matrix = [numpy_results[size]["matrix_matrix"] for size in sizes]
cupy_matrix_matrix = [cupy_results[size]["matrix_matrix"] for size in sizes]

numpy_matrix_vector = [numpy_results[size]["matrix_vector"] for size in sizes]
cupy_matrix_vector = [cupy_results[size]["matrix_vector"] for size in sizes]

sizes = [int(size) for size in sizes]  # Converting string keys to integers
```

```python
# Plotting Matrix-to-Matrix Multiplication Performance
plt.figure(figsize=(10, 5))
plt.plot(sizes, numpy_matrix_matrix, marker='o', label="NumPy")
plt.plot(sizes, cupy_matrix_matrix, marker='o', label="CuPy")
plt.xlabel("Matrix Size")
plt.ylabel("Time (s)")
plt.title("Matrix-to-Matrix Multiplication Performance")
plt.legend()
plt.grid()
plt.savefig("matrix_matrix_comparison.png")
```

```python
# Plot Matrix-to-Vector Multiplication Performance
plt.figure(figsize=(10, 5))
plt.plot(sizes, numpy_matrix_vector, marker='o', label="NumPy")
plt.plot(sizes, cupy_matrix_vector, marker='o', label="CuPy")
plt.xlabel("Matrix Size")
plt.ylabel("Time (s)")
plt.title("Matrix-to-Vector Multiplication Performance")
plt.legend()
plt.grid()
plt.savefig("matrix_vector_comparison.png")
```
